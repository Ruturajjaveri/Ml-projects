# use standard scalar to standardize amount from dataset

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
amount=df['Amount'].values
df['Amount']=sc.fit_transform(amount.reshape(-1,1))
print(df['Amount'].head(5))
```

# Perform Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop(['Class'],axis=1)
y=df['Class']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

# Knn is a supervised learning algorithm

# d=√((x2-x1)²+(y2-y1)²)


```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(x_train,y_train)
prediction=knn.predict(x_test)
```

# check accuracy,classification report,condfusin matrix

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
print(confusion_matrix(y_test, prediction))
print(classification_report(y_test, prediction))
print(accuracy_score(y_test, prediction))
```

# checking best value for k

```python
import numpy as np
error_rate=[]
for i in range(1,6):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train,y_train)
    pred=knn.predict(x_test)
    error_rate.append(np.mean(pred !=y_test))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,6),error_rate,color='blue',linestyle='dashed',marker='o',markerfacecolor='red',markersize=10)
plt.xlabel('k')
plt.ylabel('Error rate')
plt.show()
```

# change value of k as per graph to get high accuracy

```python
knn = KNeighborsClassifier(n_neighbors=2)
knn.fit(x_train,y_train)
prediction=knn.predict(x_test)
```

```python
print(confusion_matrix(y_test, prediction))
print(classification_report(y_test, prediction))
print(accuracy_score(y_test, prediction))
```

```python
from sklearn.metrics import plot_confusion_matrix
knn_cm_plot=plot_confusion_matrix(knn,x_test,y_test)
plt.show()
```
